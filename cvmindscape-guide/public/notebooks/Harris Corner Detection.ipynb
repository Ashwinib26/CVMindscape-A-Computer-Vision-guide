### CV Practical Guide (HARRIS CORNER DETECTION)

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
img = cv2.imread("D:/corner.jpg")
img_cpy= np.copy(img)
```

```python
img_cpy = cv2.cvtColor(img_cpy,cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(img,cmap='gray')
```

```python
plt.imshow(img_cpy,cmap='gray')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
gray = np.float32(gray)
```

```python
dst = cv2.cornerHarris(src=gray, blockSize=3, ksize=3, k=0.04)
```

```python
dst = cv2.dilate(dst, None)
```

```python
img[dst > 0.02 * dst.max()] = [0, 0, 255]  # corners in red
```

```python
img_cpy = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img_cpy)
plt.title("Harris Corner Detection")
plt.axis("off")
plt.show()
```

### Conclusion : 
##### 1. Due to color high contrast in sky and trees , they are being detected as corners.
##### 2. Almost all actual corners are detected.
##### 3. Some corners that are present but there does not exist gradient , that coeners are not detected
